```python
import json

arr = [json.loads(i) for i in open("tweet-json.txt")]
#data_arr = arr[:100]
```

```python
import pandas as pd
df = pd.DataFrame(arr)
df.head(50)
```

```python
df.columns
```

```python
media_val = []
df_vals = df['extended_entities'].values
for i in range(0, len(df['extended_entities'].values)):
    d = df_vals[i]
    if type(d) == dict:
        b = d["media"]
        media_val.append(b)
print(media_val)
    
```
